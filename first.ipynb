```python
import pandas as pd
import numpy as np
def load():
    return pd.read_csv('netflix_titles.csv')
def peek(df):
    return df.head()
def info(df):
    return df.info()
def clean(df):
    df = df.copy()
    df['director'] = df['director'].fillna('Unknown')
    df['cast'] = df['cast'].fillna('Unknown')
    df['country'] = df['country'].fillna('Unknown')
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
    df = df.dropna(subset=['rating', 'duration']) 
    return df
#top 10 contries with the most content
def top_countries(df):
    return (df['country'] .str.split(',').explode().str.strip().value_counts().head(10))
#average duration of movies
def average_duration(df):
    movies = df[df['type'] == 'Movie'].copy()
    movies['mins'] = movies['duration'].str.extract(r'(\d+)').astype(float)
    return movies.groupby('release_year')['mins'].mean().sort_index()
#prase dates and find peak release month
def peak_release_month(df):
    df = df.copy()
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
    return df['date_added'].dt.month.value_counts().idxmax()
#mean,median, std of movie durations
def movie_duration_stats(df):
    df = df.copy()
    df['duration'] = df['duration'].str.extract('(\d+)').astype(float)
    movies = df[df['type'] == 'Movie']
    return movies['duration'].mean(), movies['duration'].median(), movies['duration'].std()
#top 10 profilic directors with the most content on netflix
def top_directors(df):
    return df['director'].value_counts().head(10)
#most common rating
def most_common_rating(df):    
    return df['rating'].value_counts().idxmax()
    
```
