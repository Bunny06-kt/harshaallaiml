```python
# Cell 1 — Load & Peek
import pandas as pd
df = pd.read_csv("orders.csv")
print("=== First 5 Rows ===")
print(df.head())

# Cell 2 — Shape
print(f"\nRows: {df.shape[0]}, Columns: {df.shape[1]}")

# Cell 3 — Missing Values & Info
print("\n=== Missing Values ===")
print(df.isnull().sum())
print("\n=== Info ===")
df.info()

# Cell 4 — Stats
print("\n=== Statistical Summary ===")
print(df.describe())

# Cell 5 — New Column
df["TotalAmount"] = df["Quantity"] * df["Price"]
print("\n=== With TotalAmount Column ===")
print(df.head())

# Cell 6 — Shipped counts
print("\n=== Shipped vs Not Shipped ===")
print(df["Shipped"].value_counts())

# Cell 7 — Bonus
print("\n=== Country with Most Orders ===")
print(df["Country"].value_counts().idxmax())

print("\n=== Most Expensive Product ===")
print(df.loc[df["Price"].idxmax()])

print("\n=== Total Revenue ===")
print(f"${df['TotalAmount'].sum():,.2f}")

# Cell 8 — Save
df.to_csv("orders_enriched.csv", index=False)
print("\n✅ Saved to orders_enriched.csv")
```

```python
def load_data():
    return pd.read_csv("order.csv")
def peek_data(df):
    return(df.head())
def data_shape(df):
    return (df.shape[0], df.shape[1])
def missing_values(df):
    return df.isnull().sum()  
def data_info(df):
    return df.info()
def stats_summary(df):
    return df.describe()
def add_total_amount(df):
    df["TotalAmount"] = df["Quantity"] * df["Price"]
    return df
def clean_data(df):#Remove:Quantity <= 0Price <= 0
    df = df[(df["Quantity"] > 0) & (df["Price"] > 0)]
    return df
def convert_orders(df):
    df["OrderDate"] = pd.to_datetime(df["OrderDate"])
    return df
def total_revenue(df):
    return df["TotalAmount"].sum()
def top_countries(df):
    return df.groupby("Country")["TotalAmount"].sum()
def top_customers(df):
    return df.groupby("CustomerID")["TotalAmount"].sum().sort_values(ascending=False).head(5)
def montly_revenue(df):
    df["Month"] = df["OrderDate"].dt.to_period("M")
    return df.groupby("Month")["TotalAmount"].sum()
def repated_customers(df):
    return df["CustomerID"].value_counts()[df["CustomerID"].value_counts() > 1].count()
def average_per_country(df):
    return df.groupby("Country")["TotalAmount"].mean()
def top_categories_by_country(df):
    return df.groupby(["Country", "Category"])["TotalAmount"].sum().unstack().fillna(0).idxmax(axis=1)

```
