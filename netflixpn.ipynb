```python
import pandas as pd

def load():
    return pd.read_csv('netflix_titles.csv')
def peek(df):
    return df.head()
def info(df):   
    return df.info()
def describe(df): 
    return df.describe()
def isnull(df):
    return df.isnull().sum()
#movies vs tv shows
def value_counts(df, column): 
    return df[column].value_counts()
# Better approach: fill or drop selectively per task
def clean(df):
    df = df.copy()
    df['director'] = df['director'].fillna('Unknown')
    df['cast'] = df['cast'].fillna('Unknown')
    df['country'] = df['country'].fillna('Unknown')
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
    df = df.dropna(subset=['rating', 'duration']) 
    return df
def movies_only(df):
    return df[df['type'] == 'Movie']
def most_common_rating(df):# Find the most common rating
    return df['rating'].value_counts().idxmax()
#print the dataframe
def print_df(df):
    print(df)
#print("\n=== Cleaned DataFrame ===")
print_df(df)
```
