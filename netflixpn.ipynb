```python
import pandas as pd

def load():
    return pd.read_csv('netflix_titles.csv')
def peek(df):
    return df.head()
def info(df):   
    return df.info()
def describe(df): 
    return df.describe()
def isnull(df):
    return df.isnull().sum()
#movies vs tv shows
def value_counts(df, column): 
    return df[column].value_counts()
def clean(df):
    df = df.dropna(subset=['director', 'cast', 'country', 'date_added', 'rating', 'duration'])
    return df
def movies_only(df):
    return df[df['type'] == 'Movie']
def most_common_rating(df):# Find the most common rating
    return df['rating'].value_counts().idxmax()
```
